```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import seaborn as sns
```

```python

# Nastavení semene pro reprodukovatelnost
np.random.seed(0)

# Jednovýběrový t-test
sample_data = np.random.normal(loc=5.0, scale=1.0, size=30)
population_mean = 5.5
t_stat_one_sample, p_value_one_sample = stats.ttest_1samp(
    sample_data, population_mean)
print(
    f"Jednovýběrový t-test: t-stat = {
        t_stat_one_sample:.2f}, p-value = {p_value_one_sample:.2e}"
)

# Dvojvýběrový t-test (nepárový)
group_A = np.random.normal(loc=5.0, scale=1.0, size=30)
group_B = np.random.normal(loc=6.0, scale=1.0, size=30)
t_stat_ind, p_value_ind = stats.ttest_ind(group_A, group_B)
print(
    f"Dvojvýběrový t-test (nepárový): t-stat = {
        t_stat_ind:.2f}, p-value = {p_value_ind:.2e}"
)

# Párový t-test
before_treatment = np.array([200, 195, 198, 205, 202])
after_treatment = np.array([180, 190, 185, 188, 183])
t_stat_paired, p_value_paired = stats.ttest_rel(
    before_treatment, after_treatment)
print(
    f"Párový t-test: t-stat = {t_stat_paired:.2f}, p-value = {p_value_paired:.2e}")

# Vykreslení grafů
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Jednovýběrový t-test
axs[0].hist(sample_data, bins=10, color="skyblue", edgecolor="black")
axs[0].axvline(
    population_mean,
    color="red",
    linestyle="dashed",
    linewidth=2,
    label=f"Population Mean = {population_mean}",
)
axs[0].axvline(
    np.mean(sample_data),
    color="green",
    linestyle="dashed",
    linewidth=2,
    label=f"Sample Mean = {np.mean(sample_data):.2f}",
)
axs[0].set_title("Jednovýběrový t-test")
axs[0].set_xlabel("Value")
axs[0].set_ylabel("Frequency")
axs[0].legend()
axs[0].grid(True)

# Dvojvýběrový t-test (nepárový)
axs[1].hist(
    group_A, bins=10, alpha=0.5, label="Group A", color="blue", edgecolor="black"
)
axs[1].hist(
    group_B, bins=10, alpha=0.5, label="Group B", color="orange", edgecolor="black"
)
axs[1].axvline(
    np.mean(group_A),
    color="blue",
    linestyle="dashed",
    linewidth=2,
    label=f"Mean A = {np.mean(group_A):.2f}",
)
axs[1].axvline(
    np.mean(group_B),
    color="orange",
    linestyle="dashed",
    linewidth=2,
    label=f"Mean B = {np.mean(group_B):.2f}",
)
axs[1].set_title("Dvojvýběrový t-test (nepárový)")
axs[1].set_xlabel("Value")
axs[1].set_ylabel("Frequency")
axs[1].legend()
axs[1].grid(True)

# Párový t-test
axs[2].plot(before_treatment, label="Before Treatment",
            marker="o", color="blue")
axs[2].plot(after_treatment, label="After Treatment", marker="o", color="red")
axs[2].set_title("Párový t-test")
axs[2].set_xlabel("Sample")
axs[2].set_ylabel("Value")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

Vysvětlení kódu
1.	Jednovýběrový t-test:
	- Generuje data s průměrem 5.0.
	- Populační průměr je 5.5.
	- Vykresluje histogram s průměrem vzorku a populačním průměrem.
2.	Dvojvýběrový t-test (nepárový):
	- Generuje dvě skupiny dat s průměry 5.0 a 6.0.
	- Vykresluje histogramy obou skupin s jejich průměry.
3.	Párový t-test:
	- Generuje hodnoty před a po léčbě s jasným rozdílem.
	- Vykresluje čárový graf s hodnotami před a po léčbě.

Výstup
- Jednovýběrový t-test: T-statistika a p-hodnota ukazují, zda je rozdíl mezi průměrem vzorku a populačním průměrem statisticky významný.
- Dvojvýběrový t-test (nepárový): T-statistika a p-hodnota ukazují, zda je rozdíl mezi průměry dvou nezávislých vzorků statisticky významný.
- Párový t-test: T-statistika a p-hodnota ukazují, zda je rozdíl mezi průměry dvou souvisejících vzorků statisticky významný.

---

```python
# Generování náhodných dat pro případovou studii
np.random.seed(0)
before_diet = np.random.normal(loc=200, scale=30, size=30)
after_diet = before_diet - np.random.normal(
    loc=20, scale=10, size=30
)  # Předpokládáme pokles po dietě
control = np.random.normal(loc=200, scale=30, size=30)

# 1. Párový t-test: Porovnání před a po dietě
t_stat_paired, p_value_paired = stats.ttest_rel(before_diet, after_diet)
print(
    f"Párový t-test: t-stat = {t_stat_paired:.2f}, p-value = {p_value_paired:.5f}")

# 2. Dvojvýběrový t-test (nepárový): Porovnání experimentální a kontrolní skupiny po dietě
t_stat_ind, p_value_ind = stats.ttest_ind(after_diet, control)
print(
    f"Dvojvýběrový t-test (nepárový): t-stat = {
        t_stat_ind:.2f}, p-value = {p_value_ind:.5f}"
)

# 3. Jednovýběrový t-test: Porovnání kontrolní skupiny s populačním průměrem (např. 190)
population_mean = 190
t_stat_one_sample, p_value_one_sample = stats.ttest_1samp(
    control, population_mean)
print(
    f"Jednovýběrový t-test: t-stat = {
        t_stat_one_sample:.2f}, p-value = {p_value_one_sample:.5f}"
)
```

```python
# Vykreslení hustotních grafů
fig, axs = plt.subplots(1, 3, figsize=(18, 5))

# Hustotní graf pro jednovýběrový t-test
sns.kdeplot(control, ax=axs[0], label="Control", color="green", fill=True)
axs[0].axvline(
    population_mean,
    color="red",
    linestyle="dashed",
    linewidth=2,
    label=f"Population Mean = {population_mean}",
)
axs[0].axvline(
    np.mean(control),
    color="blue",
    linestyle="dashed",
    linewidth=2,
    label=f"Control Mean = {np.mean(control):.2f}",
)
axs[0].set_title("Jednovýběrový t-test")
axs[0].set_xlabel("Cholesterol Level")
axs[0].set_ylabel("Density")
axs[0].legend()
axs[0].grid(True)

# Hustotní graf pro dvojvýběrový t-test (nepárový)
sns.kdeplot(after_diet, ax=axs[1], label="After Diet", color="red", fill=True)
sns.kdeplot(control, ax=axs[1], label="Control", color="green", fill=True)
axs[1].set_title("Dvojvýběrový t-test (nepárový)")
axs[1].set_xlabel("Cholesterol Level")
axs[1].set_ylabel("Density")
axs[1].legend()
axs[1].grid(True)

# Hustotní graf pro párový t-test
sns.kdeplot(before_diet, ax=axs[2],
            label="Before Diet", color="blue", fill=True)
sns.kdeplot(after_diet, ax=axs[2], label="After Diet", color="red", fill=True)
axs[2].set_title("Párový t-test")
axs[2].set_xlabel("Cholesterol Level")
axs[2].set_ylabel("Density")
axs[2].legend()
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

### 1. Jednovýběrový t-test (Porovnání kontrolní skupiny s populačním průměrem)
- **Statistiky**: t-stat = 2.60, p-hodnota = 0.01361
- **Interpretace**: Nízká p-hodnota značí, že průměrná hladina cholesterolu v kontrolní skupině se statisticky významně liší od předpokládaného populačního průměru 190. Konkrétně, průměr kontrolní skupiny je vyšší než 190, což naznačuje, že kontrolní skupina má vyšší průměrnou hladinu cholesterolu než populační průměr.

- **Grafická interpretace**: Distribuce kontrolní skupiny (zelená) ukazuje, že její průměrná hodnota (modrá čára) je nad červenou čárou označující populační průměr. Tento vizuální výsledek je v souladu s výsledky jednovýběrového t-testu, což ukazuje významný rozdíl od populačního průměru.

### 2. Dvojvýběrový t-test (nepárový) (Porovnání experimentální a kontrolní skupiny po dietě)
- **Statistiky**: t-stat = -1.93, p-hodnota = 0.05802
- **Interpretace**: P-hodnota je těsně nad obvyklým prahem statistické významnosti (0.05), což naznačuje, že není dostatečný důkaz pro to, aby se tvrdilo, že mezi skupinami existuje statisticky významný rozdíl v hladinách cholesterolu po dietě. To může naznačovat, že jakékoli rozdíly mezi skupinami mohou být spíše náhodné.

- **Grafická interpretace**: Graf znázorňuje překrývání distribucí hladin cholesterolu pro skupinu po dietě (červená) a kontrolní skupinu (zelená). Překryv distribucí odpovídá výsledku t-testu, kde není jasně rozlišitelný rozdíl mezi skupinami.


### 3. Párový t-test (Porovnání před a po dietě)
- **Statistiky**: t-stat = 15.33, p-hodnota = 0.00000
- **Interpretace**: Výsledek párového t-testu ukazuje extrémně nízkou p-hodnotu, což indikuje, že existuje statisticky významný rozdíl v úrovních cholesterolu před a po dietě. Tento výsledek naznačuje, že dieta měla významný vliv na snížení hladiny cholesterolu u účastníků.

- **Grafická interpretace**: Na grafu jsou zobrazeny rozdělení úrovní cholesterolu před dietou (modrá) a po dietě (červená). Rozdělení po dietě je posunuto výrazně doleva oproti rozdělení před dietou, což graficky potvrzuje snížení hladin cholesterolu.
